```python
import sys, os, time
sys.path.append("../")
sys.path.append("../web_tool/")

os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "2"
os.environ['TF_CPP_MIN_LOG_LEVEL'] = "3"

import joblib

import numpy as np
```

```python
import json
```

```python
output = {
    "type": "FeatureCollection",
    "name": "hcmc",
    "crs": { "type": "name", "properties": {"name":"urn:ogc:def:crs:EPSG::3857"}},
    "features": []
}

for directory in [
    "../data/atimur_data/wjvghfqo",
    "../data/atimur_data/kdemkkmf",
    "../data/atimur_data/cumsieye"
]:
    fns = os.listdir(directory)
    fns = [
        fn
        for fn in fns
        if "request_list" in fn
    ]
    fns = sorted(fns, key=lambda x: int(x.split("_")[1]))
    print(fns[-1])
    
    request_list = joblib.load(os.path.join(directory, fns[-1]))
    
    
    for request in request_list:
        if request["type"] == "correction":

            feature = {
                "type": "Feature",
                "geometry": {
                    #"type": "Polygon",
                    "type": "Point",
                    "coordinates": None
                },
                "properties": { "user_label": request["value"]}
            }
            xmin = request["extent"]["xmin"]
            xmax = request["extent"]["xmax"]
            ymin = request["extent"]["ymin"]
            ymax = request["extent"]["ymax"]
            polygon = [[xmin,ymax],[xmax,ymax],[xmax,ymin],[xmin,ymin],[xmin,ymax]]
            #feature["geometry"]["coordinates"] = [polygon]

            point = [
                (xmin+xmax)/2,
                (ymin+ymax)/2
            ]

            feature["geometry"]["coordinates"] = point

            output["features"].append(feature)
```

```python
len(output["features"])
```

```python
with open("../data/atimur_data/outputs.geojson", "w") as f:
    f.write(json.dumps(output))
```
